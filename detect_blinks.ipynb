```python
import os
main_path = os.path.abspath('.')
```

```python
import glob
import cv2
import numpy as np
import time
# user-defined libraries
from detect_blink import detect_blinks 
from detect_blink import face_detector 
```

```python
os.chdir(main_path)
images = [cv2.imread(file) for file in glob.glob('./sample_pictures/s01/*.JPG')]
os.chdir('./detect_blink')
score_blinks = []
score_optflow = []
faces_blinks = np.zeros((len(images), 2))
blink_count = []
t1 = time.time()
# evaluate eye blinks
cnt = 0
for img in images:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.hogg_face_detector(gray)
    faces_blinks[cnt] = detect_blinks.detect_blink(img, gray, faces)
    blink_count.append(faces_blinks[cnt, 1])
    cnt += 1
t2 = time.time() - t1
print(t2)
os.chdir('..')
```
